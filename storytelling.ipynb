# Perfil das buscas por voos executivos

Os voos executivos possuem as tarifas mais altas entre os serviços de voos comerciais. Considerando a importância desse segmento no mercado, será apresentada uma análise usando os dados do log de buscas no MaxMilhas, durante o período de janeiro até março do ano de 2017, para entender melhor a demanda por buscas de passagens aéreas na classe executiva.

A primeira etapa do processo de análise consiste em filtrar os registros sobre voos executivos no log de buscas. Depois serão respondidas as questões da análise através de visualizações de dados.

```python
import main_analysis as an
import pandas as pd

df = pd.read_csv('executive.csv',low_memory=False)
```

## Quais tipos de passagens são mais procuradas? Ida ou Ida/Volta?

```python
an.countTypeOfFlights(df)
```

* A maioria das buscas são para passagens de ida/volta, mas ainda temos muitas pessoas que buscam apenas passagens de ida. Isso pode ser atribuído às altas tarifas que cobram pelo serviço, o que pode fazer com que algumas pessoas escolham ir na classe executiva e voltar na classe econômica.

## Quais são as origens e os destinos mais procurados para passagens de ida?

```python
an.topAirportsOnlyToFlights(df)
```

* A maioria das origens são de aeroportos no sudeste brasileiro. Já nos destinos, temos uma representatividade maior de aeroportos de outros países, principalmente dos Estados Unidos.

* É possível ver que aeroportos do estado de São Paulo lideram a quantidade de buscas, tanto nos destinos quantos nas origens, em passagens de apenas ida. Isso mostra que muitas pessoas que já estão fora do país tem interesse em voltar na classe executiva.

## Quais são as origens e os destinos mais procurados para passagens de ida/volta?

```python
an.topAirportsFromToFlights(df)
```

* Diferente das passagens só de ida, nesse tipo de passagem as origens mais procuradas são predominadas pelos aeroportos nacionais e os destinos mais procurados são predominados pelos aeroportos de outros países.

* A maior parte dos destinos nos voos executivos são para os Estados Unidos. 


## Quantos dias duram em média as viagens onde são procuradas passagens de ida/volta?

```python
an.averageTripDuration(df)
```

* As viagens para os destinos mais procurados duram em média de 9 até 17 dias.
* A maior duração encontrada é para o aeroporto de Barcelona na Espanha, mesmo sendo o destino com menor quantidade de buscas entre os destinos preferidos. Isso mostra que existe um grupo menor de pessoas que procuram viagens de longa duração na classe executiva para região de Barcelona no período analisado.

## Qual a relação entre as origens e os destinos preferidos nas buscas por passagens de ida/volta?

```python
an.relationTopAirports(df)
```

* O gráfico ilustra a quantidade de buscas de cada origem preferida para cada destino preferido.
* Os aeroportos do estado de São Paulo, Confins e Brasília concentram grade partes dos voos para Miami.
* Confins também concentra um número considerável das viagens para Lisboa.

## Qual a relação entre os destinos preferidos e as companhias aéreas?

```python
an.relationTopAirportsAirlines(df)
```

* O gráfico ilustra a quantidade de buscas de cada companhia aérea para cada destino preferido.
* A TAM possui o maior número de buscas para todos os destinos preferidos, com exceção da Flórida, onde provavelmente não oferece o serviço.
* A GOL tem números muito próximos dos da TAM e registra buscas para todos os destinos preferidos.
* A Azul tem quantidades de buscas bem menores em comparação com seus dois concorrentes.

## Quanto tempo dura em média a busca pelas passagens nos destinos preferidos?

```python
an.averageSearchDuration(df)
```

* As buscas para os destinos preferidos duram em média de 2 a 8 segundos.
* Os maiores tempos de busca são para os destinos nos Estados Unidos, que são justamente os destinos responsáveis pela maior parte de buscas por voos executivos. Minimizar esses tempos de busca pode melhorar a experiência do usuário.
